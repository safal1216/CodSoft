```python
import tkinter as tk
import random
import re

# Function for numeric input
def fun_num(num):
    global expression, flag, end
    if end:
        result.delete(0, tk.END)
    pre_num = result.get()
    result.delete(0, tk.END)
    result.insert(0, str(pre_num) + str(num))
    expression = result.get()
    flag = True
    end = False

# Function for calculation
def fun_cal(num):
    global flag, end
    if end:
        result.delete(0, tk.END)
    if flag:
        global expression
        pre_num = result.get()
        result.delete(0, tk.END)
        result.insert(0, str(pre_num) + str(num))
        expression = result.get()
        flag = False
        end = False

# Function for clearing the screen
def fun_clrscrn():
    result.delete(0, tk.END)

# Function for removing leading zeroes from a number to avoid octal interpretation
def remove_leading_zero(txt):
    new = ""
    count = 1
    x = re.split(r"\D+", txt)
    y = re.split("\d+", txt)
    sign_len = len(y)
    for i in range(0, len(x)):
        x[i] = str(int(x[i]))

    for i in x:
        new = new + i
        if sign_len != 1:
            new = new + y[count]
            sign_len -= 1
            count += 1
    return new

# Function for evaluating the expression
def fun_equal():
    global expression, flag, end
    expression = remove_leading_zero(expression)

    if flag:
        try:
            result.delete(0, tk.END)
            result.insert(0, eval(expression))
        except Exception as e:
            result.insert(0, str(e))
        finally:
            end = True
            flag = False

# Function to generate random RGB colors
def generate_random_color():
    r = lambda: random.randint(0, 255)
    color = '#{:02x}{:02x}{:02x}'.format(r(), r(), r())
    return color

# Create the root window
root = tk.Tk()
root.geometry("379x278")
root.title("Basic Calculator")

# Global declaration
flag = False
end = False
expression = ''

# Define different colors for the buttons
color_1 = '#33AAFF'  # Light Blue
color_2 = '#AA33FF'  # Light Purple
color_3 = '#33FF57'  # Green

# Buttons for numeric input values
button_1 = tk.Button(root, text="1", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(1))
button_2 = tk.Button(root, text="2", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(2))
button_3 = tk.Button(root, text="3", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(3))
button_4 = tk.Button(root, text="4", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(4))
button_5 = tk.Button(root, text="5", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(5))
button_6 = tk.Button(root, text="6", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(6))
button_7 = tk.Button(root, text="7", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(7))
button_8 = tk.Button(root, text="8", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(8))
button_9 = tk.Button(root, text="9", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(9))
button_0 = tk.Button(root, text="0", padx=40, pady=20, borderwidth=1, bg=color_1, command=lambda: fun_num(0))

# buttons for performing various operations
button_mul = tk.Button(root, text="*", padx=40, pady=20, borderwidth=1, bg=color_2, command=lambda: fun_cal('*'))
button_sub = tk.Button(root, text="-", padx=42, pady=20, borderwidth=1, bg=color_2, command=lambda: fun_cal('-'))
button_add = tk.Button(root, text="+", padx=40, pady=20, borderwidth=1, bg=color_2, command=lambda: fun_cal('+'))
button_divide = tk.Button(root, text="/", padx=40, pady=20, borderwidth=1, bg=color_2, command=lambda: fun_cal('/'))

# Adding clear and equal button
button_clear = tk.Button(root, text="Clear", padx=30, pady=20, borderwidth=1, bg=color_3, command=lambda: fun_clrscrn())
button_equal = tk.Button(root, text="=", padx=41, pady=20, borderwidth=1, bg=color_3, command=lambda: fun_equal())

# Entry for display result
result = tk.Entry(root, borderwidth=5, width=60)
result.grid(row=0, column=0, columnspan=30)

# Grid layout for buttons
button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)
button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)
button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)
button_0.grid(row=4, column=0)

button_mul.grid(row=3, column=3)
button_sub.grid(row=2, column=3)
button_add.grid(row=1, column=3)
button_divide.grid(row=4, column=3)
button_clear.grid(row=4, column=2)
button_equal.grid(row=4, column=1)

root.mainloop()
```
